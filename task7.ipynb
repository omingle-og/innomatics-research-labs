```python
class SmartLight:
    def __init__(self, room_name, status="OFF"):
        self.room_name = room_name
        self.status = status

    def toggle_light(self, new_status):
        self.status = new_status.upper()
        print(f"Task 1 Output: {self.room_name} Light is {self.status}")

def main_task1():
    light = SmartLight("Bedroom")
    light.toggle_light("ON")

main_task1()
```

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Task 2 Output: --- Employee ID Card ---")
        print(f"Name: {self.name} | ID: {self.emp_id} | Dept: {self.department}")

def main_task2():
    emp = Employee("Arjun Mehta", "EMP101", "Data Science")
    emp.display_id_card()

main_task2()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def show_contact(self):
        print(f"Task 3 Output: Contact Name: {self.name} | Phone: {self.phone}")

def main_task3():
    c1 = Contact("Rahul", "+91 9876543210")
    c1.show_contact()

main_task3()
```

```python
class Product:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    def print_price_tag(self):
        print(f"Task 4 Output: Product: {self.product_name} | Price: ₹{self.price}")

def main_task4():
    p1 = Product("Wireless Mouse", 2499)
    p1.print_price_tag()

main_task4()
```

```python
class Movie:
    def __init__(self, title, rating):
        self.title = title
        self.rating = rating

    def show_rating(self):
        print(f"Task 5 Output: Movie: {self.title} | Rating: {self.rating}/5")

def main_task5():
    m1 = Movie("Inception", 4.8)
    m1.show_rating()

main_task5()
```

```python
class DeliveryAddress:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_address(self):
        print(f"Task 6 Output: Delivery for: {self.customer_name} | Address: {self.address}")

def main_task6():
    d1 = DeliveryAddress("Sita Sharma", "123 Green Lane, Assam")
    d1.display_address()

main_task6()
```
