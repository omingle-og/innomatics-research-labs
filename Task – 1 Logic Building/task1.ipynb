```python
marks = [45, 78, 90, 33, 60]
passed = 0
failed = 0 
for i in marks:
    if i >= 50:
        passed += 1
    else:
        failed += 1

print("number of students passed:", passed)
print("number of students failed:", failed)
```

```python
names = [" Alice ", "bob", " CHARLIE "]
cleansed_names = []
for name in names:
    cleansed_names.append(name.strip().lower())
print(cleansed_names)
```
