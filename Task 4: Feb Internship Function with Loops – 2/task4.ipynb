```python
# Problem 1: Mobile Recharge Validation System
def validate_recharge(plan_amount):
    valid_plans = [199, 299, 399, 599]      
    if plan_amount < 50:
        return False, "Recharge amount must be at least ₹50."
    if plan_amount not in valid_plans:
        return False, f"Invalid plan. Choose from {valid_plans}."
    return True, "Recharge successful."

while True:
    try:
        amount = int(input("Enter recharge plan amount: "))
        ok, msg = validate_recharge(amount)
        print(msg)
        if ok:
            break
        retry = input("Retry? (y/n): ").strip().lower()
        if retry != "y":
            break
    except ValueError:
        print("Please enter a valid number.")
```

```python
# Problem 2: Inventory Reorder Alert System
def check_inventory(stock_dict):
    for product, qty in stock_dict.items():
        if qty < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

inventory = {
    "Rice": 20,
    "Oil": 10,
    "Sugar": 14,
    "Salt": 25
}
check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System
def student_result(marks):
    if not marks:
        return "No marks provided."
    total = 0
    for m in marks:
        total += m
    avg = total / len(marks)
    result = "Pass" if avg >= 50 else "Fail"
    return f"Average: {avg:.2f}, Result: {result}"

print(student_result([40, 55, 60, 45]))
```

```python
# Problem 4: Cab Fare Estimator with Retry Option
def calculate_fare(distance_km, peak_hour=False):
    base_fare = 50
    fare = base_fare + (12 * distance_km)
    if peak_hour:
        fare += fare * 0.25
    return fare

while True:
    try:
        km = float(input("Enter distance in km: "))
        peak = input("Peak hour? (y/n): ").strip().lower() == "y"
        total_fare = calculate_fare(km, peak)
        print(f"Estimated fare: ₹{total_fare:.2f}")
        again = input("Calculate again? (y/n): ").strip().lower()
        if again != "y":
            break
    except ValueError:
        print("Please enter valid distance.")
```

```python
# Problem 5: Employee Attendance Eligibility Checker
def attendance_eligibility(attendance_list):
    if not attendance_list:
        return "Not Eligible"
    present_days = 0
    for day in attendance_list:
        if day.upper() == "P":
            present_days += 1
    percentage = (present_days / len(attendance_list)) * 100
    return "Eligible" if percentage >= 75 else "Not Eligible"

print(attendance_eligibility(["P", "A", "P", "P", "A", "P", "P", "P"]))
```

```python
# Problem 6: Password Strength Checker
def check_password_strength(password):
    if len(password) < 8:
        return "Weak: Minimum length is 8."

    has_digit = False
    has_special = False
    special_chars = "@#$"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if not has_digit:
        return "Weak: Add at least one digit."
    if not has_special:
        return "Weak: Add at least one special character (@#$)."

    return "Strong password."

print(check_password_strength("Test@123"))
```
