Store employee names and their salaries, and determine which employee has the highest salary.


```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_salary = max(employees.values())
highest_paid_employee = [name for name, salary in employees.items() if salary == highest_salary]
for employee in highest_paid_employee:
    print(f"The highest paid employee is {employee} with a salary of {highest_salary}.")
```

Problem 5: Calculate Attendance Percentage
Problem Statement
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).



```python
attendance = ["P", "P", "A", "P", "P"]
absent_count = attendance.count("A")
present_count = attendance.count("P")
print(f"Attendance percent is {present_count / len(attendance) * 100:.2f}%")
```

Problem 9: Check if a Key Exists in a Dictionary
Problem Statement

```python
# create a  dictionary with 5 key-value pairs, where the keys are strings and the values are integers. Then, find the key with the highest value and print it.
my_dict = {
    "apple": 50,
    "banana": 30,
    "orange": 70,
    "grape": 40,
    "mango": 60
}
a = input("enter a key: ")
if a in my_dict:
    print(f"The value of key '{a}' is {my_dict[a]}")
else:
    print("Key not found")
```
