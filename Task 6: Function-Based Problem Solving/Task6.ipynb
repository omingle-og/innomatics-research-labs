```python
def manage_parking_lot(total_capacity, vehicle_logs):
    current_occupancy = sum(vehicle_logs)
    peak_occupancy = 0
    temp_count = 0

    for log in vehicle_logs:
        temp_count += log
        if temp_count > peak_occupancy:
            peak_occupancy = temp_count

    status = "ALARM: Capacity Exceeded" if current_occupancy > total_capacity else "Normal"

    return {
        "Current": current_occupancy,
        "Peak": peak_occupancy,
        "Status": status
    }

print("Problem 1:", manage_parking_lot(50, [20, 35, -10, 15]))
```

```python
def estimate_delivery(distance_km, weather, traffic):
    base_time = 10
    time_per_km = 3

    weather_delay = {"Clear": 0, "Rainy": 10, "Stormy": 20}
    traffic_delay = {"Low": 0, "Medium": 5, "High": 15}

    total_eta = base_time + (distance_km * time_per_km) + weather_delay.get(weather, 0) + traffic_delay.get(traffic, 0)

    return f"Estimated Delivery: {total_eta} mins"

print("Problem 2:", estimate_delivery(5, "Rainy", "High"))
```

```python
def check_theatre_status(total_seats, booked_seats):
    occupancy = (booked_seats / total_seats) * 100
    status = "Housefull" if occupancy >= 100 else "Available"

    return f"Occupancy: {occupancy:.1f}% | Status: {status}"

print("Problem 3:", check_theatre_status(200, 195))
```

```python
def classify_server_load(cpu_usage_list):
    avg_load = sum(cpu_usage_list) / len(cpu_usage_list)

    if avg_load < 50:
        return f"Avg Load: {avg_load}% - Status: Normal"
    elif 50 <= avg_load <= 80:
        return f"Avg Load: {avg_load}% - Status: Warning"
    else:
        return f"Avg Load: {avg_load}% - Status: Critical"

print("Problem 4:", classify_server_load([85, 90, 78, 92]))
```

```python
def monitor_resources(resource_dict, limit=100):
    overused = [res for res, hours in resource_dict.items() if hours > limit]

    if overused:
        return f"Alert! Overused resources: {', '.join(overused)}"
    return "All resources within limits."

usage = {"Projector": 120, "AC": 80, "Lights": 150}
print("Problem 5:", monitor_resources(usage))
```

```python
def manage_registration(capacity, current_total):
    if current_total >= capacity:
        return "Registration Full. Moving to Waitlist Mode."
    else:
        remaining = capacity - current_total
        return f"Registration Open. {remaining} slots left."

print("Problem 6:", manage_registration(100, 102))
```
